## XPath语法和xml模块

```python
from lxml import etree
```

### 方法一

```python
parser = etree.HTMLParser(encoding="UTF-8")
```

```python
html = etree.parse("tencent.html", parser)
```

```python
trs = html.xpath("//div")
for tr in trs:
    print(etree.tostring(tr,encoding="UTF-8").decode("UTF-8"))
```
