### ascii展示对象

```python
class Student():
    def __init__(self,id,name):
        self.id = id
        self.name = name
    def __repr__(self):
        return 'id = '+self.id +', name = '+self.name
```

```python
xiaoming = Student(id='1',name='xiaoming')
xiaoming
```

```python
# ascii() 函数类似 repr() 函数, 返回一个表示对象的字符串
print(ascii(xiaoming))
```

### 类方法

```python
class Student():
    name_master = "my_zev"
    def __init__(self,id,name):
        self.id = id
        self.name = name
    def __repr__(self):
        return 'id = '+self.id +', name = '+self.name
    @classmethod
    def f(cls, name_master):
        print('类方法可以调用：%s' % name_master)
```

```python
Student.f("name_master")
```

### 创建property属性的两种方式

```python
class c:
    def __init__(self):
        self._x = None
    @property
    def x(self):
        return self._x
    @x.setter
    def x(self, value):
        self._x = value
    
    @x.deleter
    def x(self):
        del self._x
```

```python
cs = c()
cs.x = 12
```

```python
del cs.x 
```

```python
cs.x =23
```

```python
cs.x
```

### 对象序列化

```python
class Student():
    def __init__(self, **args):
        self.ids = args["ids"]
        self.name = args["name"]
        self.address = args["address"]
xiaoming = Student(ids=1,name="zev",address="cd")
xiaohong = Student(ids=2,name="hy",address="hn")
```

```python
import json

with open('json.txt',"w") as f:
    json.dump([xiaoming,xiaohong], f, default=lambda obj:obj.__dict__,ensure_ascii=False)
```
