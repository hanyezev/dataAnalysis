### 缺失值填充

```python
# k近邻法
def knn_mean(ts, n):
    out = np.copy(ts)
    for i, val in enumerate(ts):
        if np.isnan(val):
            n_by_2 = np.ceil(n / 2)
            lower = np.max([0, int(i - n_by_2)])
            upper = np.min([len(ts) + 1, int(i + n_by_2)])
            ts_near = np.concatenate([ts[lower:i], ts[i:upper]])
            out[i] = np.nanmean(ts_near)
    return out
# 使用k-近邻法填补缺失值
df["主机CPU平均负载"] = knn_mean(df["主机CPU平均负载"], 24)
```

### 建立LSTM模型

```python
class lstm(nn.Module):
    def __init__(self,input_size,hidden_size,output_size,num_layer):
        super(lstm,self).__init__()
        self.rnn = nn.LSTM(input_size,hidden_size,num_layer,batch_first=True)
        self.fc = nn.Linear(hidden_size*15,output_size)
    
    def forward(self,x):
#         x,_ = self.rnn(x)
#         s,b,h = x.size()
#         x = x.view(s*b,h)
#         x = self.fc(x)
#         x = x.view(s,b,-1)
        x, (h, c) = self.rnn(x)
        x = x.reshape(256, -1)
        x = self.fc(x)  # [seq, h] => [seq, 1]
        return x

model = lstm(1, 10, 1, 1)
model
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
epoch_n = 20
```

```python
losses =[] 
lr_list = []
for epoch in range(epoch_n):
    running_loss = 0.0
    for step, data in enumerate(train_data, 0):
        x, y = data
        y.reshape(256)
        output = model(x).squeeze(1)
        loss = criterion(output, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
#         for p in model.parameters():
#             # 查看每个梯度的模
#             print(p.grad.norm())
#             torch.nn.utils.clip_grad_norm_(p, 10)
        running_loss += loss.item()
        if step % 2 == 1:    # print every 2 mini-batches
            print('[%d, %d] loss: %.7f' % (epoch + 1, step + 1, running_loss/2))
#             ep.append(f'{epoch + 1}-{step + 1}')
            losses.append(running_loss)
            running_loss = 0.0
#     if (epoch+1)%20 == 0:
#         for p in optimizer.param_groups:
#             p['lr'] *= 0.1
#     lr_list.append(optimizer.state_dict()['param_groups'][0]['lr'])
print('Finished Training')

# 开始训练
# for e in range(100):
#     var_x = Variable(train_X)
#     var_y = Variable(train_Y)
#     # 前向传播
#     out = model(var_x)
#     loss = criterion(out, var_y)
#     # 反向传播
#     optimizer.zero_grad()
#     loss.backward()
#     optimizer.step()
    
#     if (e + 1) % 5 == 0: # 每 2 次输出结果
#         print('Epoch: {}, Loss: {:.5f}'.format(e + 1, loss.item()))
# print('Finished Training') 
```
